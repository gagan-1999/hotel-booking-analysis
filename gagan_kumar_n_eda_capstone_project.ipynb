#**Removing missing values**


```python
#check for missing/null values in Dataset
null_df=pd.DataFrame(df.isna().sum()).rename(columns={0:'number_of_nulls_values'})
null_df
```

```python
#percentage of null values
percentage_null_df=pd.DataFrame(round(df.isna().sum()*100/len(df),4)).rename(columns={0:'percentage_null_values'})
percentage_null_df
```

Here variables children having only 4 null values. Variables country, agent and company having 0.52%, 13.95% and 93.98% null values respectively

```python
#checking category of features whoes having null values
df.country.value_counts()
```

```python
df.agent.value_counts()
```

```python
df.children.value_counts()
```

```python
#Variable having more than 50% null values will be removed
df.drop(columns='company', inplace=True)
```

```python
#Treating null vlues of descrete variable children, country and agent
df['children']=df['children'].fillna(df['children'].mode()[0])
df['country']=df['country'].fillna(df['country'].mode()[0])
df['agent']=df['agent'].fillna(df['agent'].mode()[0])
```

From above children, country and agent are descrete variable so replaced null values with mode and variable company having null values more than 50% so removed it.



```python
#Check for null values
df.isna().sum()
```

#**Handling Outliers**

```python
numerical_variable=[i for i in df.columns if df[i].dtypes!='O']
print(f'Numerical variables are', len(numerical_variable))
print('--'*50)
print(numerical_variable)
```

```python
for i in numerical_variable:
  plt.subplot(1,2,1)
  fig=sns.boxplot(y=df[i])
  fig.set_title('Box plot')
  fig.set_ylabel(i)

  plt.subplot(1,2,2)
  fig=sns.distplot(df[i])
  fig.set_title('Distribution plot')
  fig.set_xlabel(i)
```
